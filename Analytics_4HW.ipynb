##2.1
Изучите распределение домов от наличия вида на набережную



```python
data = df[ 'waterfront' ].value_counts()

plt.pie( data, autopct='%.3f%%', labels = [ 'С видом на набережную', 'Без вида на набережную' ] )
plt.title( 'Распределение домов' );
plt.legend( [ f"С видом: { data[ 0 ] }", f"Без вида: { data[ 1 ] }" ] )
```

##2.2
Изучите распределение этажей домов

```python
data=df['floors'].value_counts()
plt.figure(figsize=(5,5))
sns.barplot(x = data.index, y = data.values)
plt.title('Распределение домов по этажности')
plt.xlabel('Этажи')
plt.ylabel('Количество');
```

## 2.3
Изучите распределение состояния домов

```python
data = df[ 'condition' ].value_counts()

plt.figure( figsize=( 7, 5 ) )
sns.barplot( x = data.index, y = data )
plt.title( 'Состояние домов' )
plt.xlabel( 'Оценка состояния дома' )
plt.ylabel( 'Количество' );
```

# Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.5] = 0
sns.heatmap(corr_matrix, linewidths=0.7, cmap='coolwarm', annot = True, vmin=-1, vmax=1, center= 0);
```

```python
sns.boxplot(x=df['grade'], y=df['price'], showfliers=False)
plt.xlabel('Качество конструкции и дизайн здания')
plt.ylabel('Цена');
```

```python
sns.boxplot( x=df[ 'price' ], y=df[ 'view' ].astype( 'str' ), whis=5 );
plt.xlabel( 'price' )
plt.ylabel( 'view' )
plt.title( 'Зависимость цены от качества вида (за окном)' );
```

```python
print("Влияние жилой площади")
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

```python
sns.jointplot( x=df[ 'price' ], y=df[ 'bathrooms' ], kind='reg' );
"Зависимость числа комнат от смоимости"
```

## Выводы:
- Цена у домов с плохой конструкцией и некачественным дизайном невысокая, а чем лучше дизайн и конструкция, тем цена выше;
- Чем дороже дом, тем больше ванных комнат в нем;
- Чем выше класс жилья, тем выше его стоимость;
- Чем лучшее вид из дома, тем выше его стоимость;
- Чем больше площадь, тем дороже жилье.
